```python
import csv

csvfile = open('nycPercentSorted.csv', newline='')
csvAvg = open('nycFreqSorted.csv',newline='')

c = csv.reader(csvfile)
d = csv.reader(csvAvg)

average=[0] * 71  # change to higher value to make it non-redundant
i=0

for row in d:
    if row[0] == 'TOTAL'or row[0]=='Category' or row[0]=='year' or row[0]=='month':
        continue
    else:
       j=0
       for month in range (1,181):
           if float(row[month]) !=0:    
               j+=1
               average[i] += float(row[month])    # maintain grand average for scale later 
       average[i] /= j
       i+=1
      
percentages = []
totals = []
categories =[]
i = 0
monYear = []

for row in c:
    
    if row[0] == 'TOTAL'or row[0]=='Category' or row[0]=='year':
        continue
    elif row[0]=='month':
        for month in range(1,181):
            monYear.append(float(row[month]))
    else:
       temp = []
       categories.append(row[0])
       
       for month in range (1,181):
           if float(row[month]) !=0:    
               j+=1
           
           temp.append(float(row[month]))   # add monthly percentages of a particular crime to  list 
       
       percentages.append(temp)   # add monthly percentages list to a list 
       i += 1

predictions=[]

def predictOverYears(start,end):           # fn to give prediction for one year mentioned by start and end
    predictAyear = []
    for idx in range (0,i):
        crime = percentages[idx]
        max = 0 
        index = -1
        for it in range (start,end):     # find max and avg (exclude maximal val while computing avg) for that year            
            if crime[it]>max:
               max=crime[it]
               index = it
    
        avg = 0 

        for it in range (start,end):
            if it == index:
                continue
            else:
                avg+=crime[it]
    
        avg = float(avg/11)
        if avg != 0 :
           likeness = ((float(max - avg)/avg) * 100 )    # measure likelihood of prediction for that year 
        else :
           likeness=0
    
        temp = [likeness,monYear[index]]
        predictAyear.append(temp)            # append likelihood and index of month predicted  
    return predictAyear

yearsToTrain = 14    # change s value to no of years to be used for training 

for j in range(0,yearsToTrain):            
    n=j*12
    predictions.append(predictOverYears(n,n+12))

predictionPerCrime = []

for idx in range(0,i):
    tem = []
    for k in range(0,yearsToTrain):
        tem.append(predictions[k][idx])           # get each individual year prediciton and add it to a new list 
    predictionPerCrime.append(tem)
    
# print("PerCrimePrediction:",predictionPerCrime[0:i])

grandPrediction = []
for idx in range(0,i):
    temG = []
    for j in range(1,13):
        sum=0
        f=0
        for k in range(0,yearsToTrain):
            if predictionPerCrime[idx][k][1] == j:
                sum+=predictionPerCrime[idx][k][0]
                f+=1
        elm = [sum/yearsToTrain,j,f] # store cumulative sum of confidence rate/yearToTrain , predicted month , freq of prediction
        # elm = [sum,j,f]
        if f!= 0:
            temG.append(elm)
    grandPrediction.append(temG)


for idx in range(0,i):
    print(categories[idx],"With overall avg:",average[idx],"predictions:")
    n = len(grandPrediction[idx])
    for j in range(0,n):
        if grandPrediction[idx][j][0]!=0 :
            print("      Month:",grandPrediction[idx][j][1],"With Confidence:",grandPrediction[idx][j][0],"with freq of prediction:",grandPrediction[idx][j][2])



```
