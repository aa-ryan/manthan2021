```python
import csv

csvfile = open('check1.csv', newline='')

c = csv.reader(csvfile)

percentages = []
totals = []
categories =[]
average=[0] * 68
i = 0

for row in c:
    
    if row[1] == 'TOTAL':
        for tot in range(2,38):
            totals.append(float(row[tot]))  # add totals of monthly crime to a list
    elif row[1] == 'Category':
        continue
    else:
       temp = []
       categories.append(row[1])
       j=0
       for month in range (2,38):
           if float(row[month]) !=0:    
               j+=1
               average[i] += float(row[month])    # maintain grand average for scale later 
           
           temp.append(float(row[month]))   # add monthly frequncies of a particular crime to a list 
       
       average[i] /= j
       percentages.append(temp)   # add monthly freq list to a list 
       
       i += 1

# print("this is i:",i)
# print("this is average",average[0:68])
# print("totals:",totals[0:36])

final = []

for idx in range(0,i):
    # print(categories[idx])
    tem=[]
    tem = percentages[idx] 

    for it in range(0,36):
        if totals[it]  != float(0) :
            tem[it] = (tem[it]/totals[it]) * 100    # iterate over each monthly % and append to list appropriately
        else:
            tem[it] = float(0)
    
    final.append(tem)
    # print("this is ",idx,final[idx])


predictions=[]

for idx in range (0,i):      # applicable for a year of data *REMEMBER TO MAKE FUNCTION FROM THIS* 
    crime = final[idx]
    max = 0 
    index = 0
    for it in range (0,12):             # find max and avg (exclude maximal val while computing avg) for that year            
        if crime[it]>max:
            max=crime[it]
            index = it
    
    avg = 0 

    for it in range (0,12):
        if it == index:
            continue
        else:
            avg+=crime[it]
    
    avg = float(avg/11)
    if avg != 0 :
      likeness = ((float(max - avg)/avg) * 100 )    # measure likelihood of prediction for that year 
    else :
        likeness=0
    
    temp = [likeness,12-index]
    predictions.append(temp)            # appnd likelihood and month of prediction 
    print(idx,"Likeliness of ",categories[idx],predictions[idx],"and average is:",average[idx])


```
